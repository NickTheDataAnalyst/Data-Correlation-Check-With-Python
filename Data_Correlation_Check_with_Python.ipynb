```python
#Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading data

df = pd.read_csv(r"C:\Users\User\Downloads\movies.csv")
df
```

```python
# Looping to check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Removing NaN for missing data

df = df.dropna()
df
```

```python
# Changing data type

df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df
```

```python
# Checking for any Outliers

df.boxplot(column=['gross'])
```

```python
#Removing duplicates

df = df.drop_duplicates()
df
```

```python
# Visualizing with a scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
```

```python
# Ploting budget vs gross using seaborn

sns.regplot(x ='budget', y ='gross', data=df)
```

```python
# Making it more visual

sns.regplot(x ='budget', y ='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation check

df = df.corr(numeric_only = [True])
```

```python
df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
